3.2 Matplotlib


importamos matplotlib.pyplot pois estamos importando apenas um subpacote com utilitários para plotagem

```python
import matplotlib.pyplot as plt

dataset1 = anscombe[anscombe['dataset'] == 'I']
dataset2 = anscombe[anscombe['dataset'] == 'II']
dataset3 = anscombe[anscombe['dataset'] == 'III']
dataset4 = anscombe[anscombe['dataset'] == 'IV']

plt.plot(dataset1['x'], dataset1['y'])
```

```python
plt.plot(dataset1['x'], dataset1['y'], 'o')
```

Por padrão o plt.plot desenha linhas, se quiser mudar para pontos basta passar o parâmetro 'o'

```python
fig = plt.figure()

axesI = fig.add_subplot(2,2,1)
axesII = fig.add_subplot(2,2,2)
axesIII = fig.add_subplot(2,2,3)
axesIV = fig.add_subplot(2,2,4)
```

Estamos criando uma subplotagem, para que não seja nescessario criar uma plotagem diferente para um dos subconjuntos

```python
axesI.plot(dataset1['x'], dataset1['y'], 'o')
axesII.plot(dataset2['x'], dataset2['y'], 'o')
axesIII.plot(dataset3['x'], dataset3['y'], 'o')
axesIV.plot(dataset4['x'], dataset4['y'], 'o')
```

adicionamos subconjuntos de dados nas plotagens

```python
#acrescenta um título para cada plotagem
axesI.set_title("dataset_1")
axesII.set_title("dataset_2")
axesIII.set_title("dataset_3")
axesIV.set_title("dataset_4")

#define um rotulo para os eixos
for ax in [axesI, axesII, axesIII, axesIV]:
    ax.set_xlabel('Eixo X')
    ax.set_ylabel('Eixo Y')

#adiciona um título para toda a figura
fig.suptitle("Anscombe Data")

#ajeita os elementos de todo o layout
fig.tight_layout()

fig
```

3.3 Gráficos estatísticos usando a matplotlib

Dados de gorjeta provenientes da biblioteca seaborn

```python
tips = sns.load_dataset("tips")
print(tips)
```

Univariado = uma única variável

3.3.1.1 Histogramas

```python
#cria figura e eixos
fig2 = plt.figure()
axes1 = fig2.add_subplot(1,1,1)

#craia o histograma
axes1.hist(tips['total_bill'], bins=10)

#adiciona título e rótulos
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig2.show()
```

Bivariado = duas variáveis

3.3.2.1 Gráfico de dispersão

Scatterplots, são usados quando uma variável contínua é plotada em relação a outra variável contínua

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(tips['total_bill'], tips['tip'])

axes1.set_title('Scatterplot of Total Bill vs Tip ')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

scatter_plot.show()
```

3.3.2.2 Gráfico de caixa

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot(
    #o primeiro argumento de um gráfico de caixa é o dado,
    #pois estamos plotando várias porções de dados
    #temos que colocar cada porção em uma lista
    [tips[tips['sex'] == 'Female']['tip'],
     tips[tips['sex'] == 'Male']['tip']],
    #foi difícil mas consegui entender, o tips[tips['sex] == 'Female'] o tips da frente isola todos os sex female em um único dataframe p mesmo vale para o 'Male'

        #podemos passar um parâmetro labels opcional
        #para especificar um rótulo aos dados que passamos
     labels=['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')

boxplot.show()

```

3.3.3 Dados multivariados

É possivel plotar dados multivariados, porém pode ser complicado e muitas vezes confuso. 

Cada situação usará de diferentes formas para se plotar, mas deve tomar cuidado e atenção para ter certeza de que os dados vão estar claros

A metplotlib e o seaborn tem o próprio conjunto de paletas de corres para diferenciar os dados

```python
#cria uma variável de cor baseada no sexo

def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
    
tips['sex_color'] = tips['sex'].apply(recode_sex)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],

    #define o tamanho dos pontos com base no tamanho dos grupos
    #multiplicamos os valores por 10 para deixar os pontos maiores e enfatizar as diferenças 
    s=tips['size'] * 10,

    #define a cor do sexo
    c=tips['sex_color'],

    #define o valor de alpha para que os pontos sejam mais transparentes
    alpha=0.5
)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

scatter_plot.show()
```

3.4.1.4 Plotagem de contadores (plotagem de barras)

A plotagem de barras é usada com o intuito de contar variáveis discretas

3.4.2.1 Gráficos de dispersão

regplot vem junto uma linha de regressão 

```python
scatter, ax = plt.subplots()

#se passar o parâmetro fit_reg=False teremos apenas o gráfico de dispersão
ax = sns.regplot(data=tips, x='total_bill', y='tip', fit_reg=False)

ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```

Também é possível usar lmplot para criar um gráfico de dispersão. A diferença é que lmplot cria uma figura, enquanto regplot cria eixos

```python
fig = sns.lmplot(data=tips, x='total_bill', y='tip')
plt.show()
```

O joint plot cria um gráfico de dispersão com uma plotagem univariada em cada eixo

```python
joint = sns.jointplot(data=tips, x='total_bill', y='tip')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')

#adiciona um titulo e regula o texto pra não ficar em cima do gráfico
joint.figure.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

3.4.3.3 Facetas

As facetas serve para fazer várias plotagens sobre uma variável de categoria.

Elas facilitam a vida, ao invés de ter que criar varios subconjuntos e definir os eixos ela faz isso de forma automática.

```python
anscombe_plot = sns.lmplot(data=anscombe, x='x', y='y',
                           fit_reg=False,
                           col='dataset', col_wrap=2)
```

Nem toda função de plotagem terá os parâmetro col e col_wrap, eles servem apenas para as plotagens em figuras, e a maioria são em eixos

E ao invés disso, nós usamos FacetGrid que saberá sobre qual variável criar a faceta, e fornecer o código da plotagem individual de cada faceta

```python
facet = sns.FacetGrid(tips, col='time')
#para cada valor de tempo plota um histograma do total da conta
facet.map(sns.histplot, 'total_bill', kde=True)
facet.map(sns.rugplot, 'total_bill')
facet.set(xlim=(0, 60), ylim=(0, 40))
plt.show()
```

```python
facet = sns.FacetGrid(tips, col='day', hue='sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()

plt.show()
```

```python
fig = sns.lmplot(data=tips, x='total_bill', y='tip', fit_reg=False,
                 hue='sex', col='day')

plt.show()
```

Outra coisa que podemos fazer é ter uma variável com faceta no eixo X e outra no eixo y passando o parâmetro row

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex',)
facet.map(plt.scatter, 'total_bill', 'tip')

plt.show()
```
